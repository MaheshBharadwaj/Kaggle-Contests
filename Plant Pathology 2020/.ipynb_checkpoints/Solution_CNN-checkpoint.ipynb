# Loading DataFrames

```python
train_df = pd.read_csv(path+'train.csv')
test_df = pd.read_csv(path+'test.csv')

print('Training DataFrame:\n', train_df.head())
print('\nTest DataFrame:\n', test_df.head())
```

```python
n_healthy = train_df[train_df['healthy'] == 1].shape[0]
n_multi = train_df[train_df['multiple_diseases'] == 1].shape[0]
n_rust =  train_df[train_df['rust'] == 1].shape[0]
n_scab = train_df[train_df['scab'] == 1].shape[0]

print('Healthy: %d\nMulti: %d\nRust: %d\nScab: %d' % (n_healthy, n_multi, n_rust, n_scab))
```

```python
try:
    X = np.load('X.npy')
    y = np.load('y.npy')

except FileNotFoundError:    
    X = []
    y = []

    for i in tqdm(range(n_train+1)):
        img = cv2.imread(path+'images/Train_'+str(i)+'.jpg')
        img = cv2.resize(img, (150, 100))
        X.append(img)
        X.append(cv2.flip(img, 0))
        l = train_df.iloc[i][['healthy', 'multiple_diseases', 'rust', 'scab']].to_list()
        y.append(np.array(l))
        y.append(np.array(l))    
    X = np.array(X) / 255
    y = np.array(y)
    
    np.save('X', X)
    np.save('y', y)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,  X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size = 0.25)
```

# Training Convolutional Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Dense,Dropout
from sklearn.metrics import classification_report
```

```python
classifier = Sequential()
classifier.add(Conv2D(32, (4, 4), input_shape = (100, 150, 3), activation = 'relu', strides=(1,1)))
classifier.add(MaxPooling2D(pool_size = (3, 3)))
classifier.add(Conv2D(32, (3, 3), activation = 'relu', strides=(1, 1)))
classifier.add(MaxPooling2D(pool_size = (2, 2)))
classifier.add(Dropout(rate = 0.02))
classifier.add(Conv2D(32, (3, 3), activation = 'relu', ))
classifier.add(MaxPooling2D(pool_size = (2, 2)))
classifier.add(Flatten())
classifier.add(Dense(units = 64, activation = 'relu', kernel_regularizer=regularizers.l2(0.01)))
classifier.add(Dense(units = 64, activation = 'relu', kernel_regularizer=regularizers.l2(0.01)))
classifier.add(Dense(units = 64, activation = 'relu', kernel_regularizer=regularizers.l2(0.01)))
classifier.add(Dense(units = 4, activation = 'softmax'))
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
classifier.summary()
```

```python
history = classifier.fit(X_train, y_train, batch_size = 64, epochs = 50, validation_data = (X_test,y_test),verbose = 2, initial_epoch=10)
```

## Upto 50

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='lower right')
plt.show()
```

```python
# Plot history: Loss
plt.plot(history.history['loss'], label='Training data')
plt.plot(history.history['val_loss'], label='Validation data')
plt.title('Loss')
plt.ylabel('Loss value')
plt.xlabel('No. epoch')
plt.legend(loc="upper left")
plt.show()
```

```python
y_pred = classifier.predict(X_test)
y_pred = np.argmax(y_pred,axis = 1)
y_truth = np.argmax(y_test,axis = 1)
print(classification_report(y_truth, y_pred, target_names=['A', 'B','C', 'D']))
```
